```python
import pandas as pd
import pprint
import os
import numpy as np
from konlpy.tag import Komoran
from konlpy.tag import Twitter
import time
import pandas as pd
import re
```

```python
file_list = os.listdir('./dataset/food')
file_list,len(file_list)
```

```python
df1 = pd.read_csv('./dataset/food/'+file_list[0]
                   , engine='python'                   
                   , names = ['name','url','review']
                   , encoding='utf-8'                   
                   , na_filter =True                   
                  )
df2 = pd.read_csv('./dataset/food/'+file_list[1]
                   , engine='python'                   
                   , names = ['name','url','review']
                   , encoding='utf-8'                   
                   , na_filter =True                   
                  )
df_merge=pd.concat([df1,df2],
                               axis=0,   
                               ignore_index=True)

for idx, fl in enumerate(file_list[2:]):    
    print(fl)
    df = pd.read_csv('./dataset/food/'+fl
                   , engine='python'                   
                   , names = ['name','url','review']
                   , encoding='utf-8'                   
                   , na_filter =True                   
                  )
    df_merge=pd.concat([df_merge,df],
                                   axis=0,
                                   ignore_index=True)
```

```python
df_merge
```

```python
#df_merge=df_merge.drop('NaN')
#df_merge=df_merge.drop('nan')
#df_merge=df_merge.drop('None')
df_merge=df_merge.drop(0)
#df_merge.reset_index
df_merge['review'][2]
```

```python
type(df_merge['review'][1])
```

```python
token_doc = ''
for text in df_merge['review']:
    token_doc += re.sub('[\'|'']','',text)
```

```python
type(token_doc)
```

```python
token_doc=token_doc.split(',')
```

```python
type(token_doc),len(token_doc),token_doc[:100]
```

```python
%matplotlib inline
import nltk
```

```python
ko = nltk.Text(token_doc)   # 이름
```

```python
ko.similar('탕')
```

```python
len(ko.vocab())
```

```python
from gensim.models import Word2Vec
embedding_model = Word2Vec(token_doc, 
                                              size=10000, 
                                              window = 5,    
                                              min_count=50, 
                                              workers=4, 
                                              iter=1000, 
                                              sg=1)
```

```python
embedding_model
```

```python
print(embedding_model.most_similar(('계륵')))
```
