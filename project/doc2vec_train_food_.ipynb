```python
import pandas as pd
import pprint
import os
import numpy as np
from konlpy.tag import Komoran
from konlpy.tag import Twitter
import time
import pandas as pd
```

```python
file_list = os.listdir('./dataset/food')
file_list,len(file_list)
```

```python
index = ['A','B','C','D',
             'E','F','G','H',
             'I','J','K','L',
             'M','N','O','P']
df1 = pd.read_csv('./dataset/food/'+file_list[0]
                   , engine='python'                   
                   , names = ['name','url','review']
                   , encoding='utf-8'                   
                   , na_filter =True                   
                  )
df1['name']=index[0]
df2 = pd.read_csv('./dataset/food/'+file_list[1]
                   , engine='python'                   
                   , names = ['name','url','review']
                   , encoding='utf-8'                   
                   , na_filter =True                   
                  )
df2['name']=index[1]
df_merge=pd.concat([df1,df2], axis=0)

for idx, fl in enumerate(file_list[2:]):    
    print(fl)
    df = pd.read_csv('./dataset/food/'+fl
                   , engine='python'                   
                   , names = ['name','url','review']
                   , encoding='utf-8'                   
                   , na_filter =True                   
                  )
    df['name']=index[idx+2]
    df_merge=pd.concat([df_merge,df],axis=0)
```

```python
df_merge
```

```python
 df_merge=df_merge.drop('NaN')
df_merge
```

```python
from collections import namedtuple
TaggedDocument = namedtuple('TaggedDocument', 'words tags')
```

```python
tagged_train_docs = [TaggedDocument(d, c) for d, c in df_merge[['review', 'name']].values]
print(len(tagged_train_docs))
```

```python
from gensim.models import doc2vec
```

```python
"""
doc_vectorizer = doc2vec.Doc2Vec(size=500, 
                                 alpha=0.025,
                                 min_alpha=0.025, 
                                 seed=1234)
doc_vectorizer.build_vocab(tagged_train_docs)
"""
doc_vectorizer = doc2vec.Doc2Vec(dm=0,
                                                    size=500, 
                                                    window=8, 
                                                    alpha=0.025,
                                                    min_alpha=0.025, 
                                                    seed=1234)
doc_vectorizer.build_vocab(tagged_train_docs)
```

```python
doc_vectorizer.corpus_count, doc_vectorizer.iter
```

```python
tokens=[]
tokens.append([ t for d in df_merge['review'] for t in d])
tokens=sum(tokens,[])
```

```python
import nltk
text = nltk.Text(tokens, name='NMSC')

print(len(text.tokens),type(text.tokens))
ex_list = list(set(text.tokens))
print(len(ex_list))
```

```python
print(text.vocab().most_common(10))
```

```python
print(text.vocab().most_common()[:-20:-1])
```

```python
for epoch in range(30):
    print(epoch)
    doc_vectorizer.train(tagged_train_docs, 
                         total_examples=doc_vectorizer.corpus_count, 
                         epochs=doc_vectorizer.iter)
    doc_vectorizer.alpha -= 0.002 # decrease the learning rate
    doc_vectorizer.min_alpha = doc_vectorizer.alpha # fix the learning rate, no decay
```

```python
texts=['김밥 맛잇어 김 단무지 햄 '.split(' '),
       '바다 산 나들이 가족 여행 휴가'.split(' '),
       '고소 기름 된장 쌈장 상추 쌈 회식 고기 육즙 '.split(' ')
       ]       
print(texts)
for txt in texts:
    print(doc_vectorizer.docvecs.most_similar(positive=[doc_vectorizer.infer_vector(txt)],topn=1))
```
