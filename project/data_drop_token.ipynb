```python
import json
import pandas as pd
import numpy as np 
import codecs 
import pandas as pd
import numpy as np
from konlpy.tag import Komoran
from konlpy.tag import Twitter
from time import time
import pickle
import os
```

```python
twitter = Twitter()
def tokenizer_twitter_morphs(doc):
    return twitter.morphs(doc)

def tokenizer_twitter_noun(doc):
    return twitter.nouns(doc)

def tokenizer_twitter_pos(doc):
    return twitter.pos(doc, norm=True, stem=True)

komoran = Komoran()
def tokenizer_noun(doc):
    return komoran.nouns(doc)

def tokenizer_morphs(doc):
    return komoran.morphs(doc)
```

```python
import re
import unicodedata
def remove_tag(content):    
    cleanr =re.compile('\t|\r|\n|,')
    cleantext = unicodedata.normalize('NFKC', (re.sub(cleanr, ' ', str(content))))
    
    return re.sub('\u200b*','',cleantext)
```

```python
file_list = os.listdir('./dataset/data')
file_list
```

```python
for fs in file_list:
    df = pd.read_csv('./dataset/data/'+fs                       
                     , engine='python'                   
                     , names = ['name','url','review']
                     , encoding='utf-8'
                     , na_filter =True                   
                    )
    drop_df=df.drop_duplicates(['url'])
    print(fs)
    for idx, text in enumerate(drop_df['review']):
        if len(str(text)) < 200 :
            drop_df=drop_df.drop(idx)
        elif len(str(text)) > 10000 :
            drop_df=drop_df.drop(idx)
        else:            
            drop_df['review'][idx]=tokenizer_morphs(remove_tag(text))
            
    print('drop')
    #drop_df['review'].apply(tokenizer_morphs)
    
    drop_df.to_csv('./dataset/food/'+fs)
    print('save')
```
