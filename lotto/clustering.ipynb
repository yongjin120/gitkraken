```python
df1 = pd.read_csv('lotto.csv'
                   , engine='python'                   
                   , names = ['1','2','3','4','5','6','7']         
                  )
```

```python
N1 = np.array(df1['1'])
N2 = np.array(df1['2'])
N3 = np.array(df1['3'])
N4 = np.array(df1['4'])
N5 = np.array(df1['5'])
N6 = np.array(df1['6'])
N7 = np.array(df1['7'])
```

```python
ys = [y for y in range(1,813)]
fig = matplotlib.pyplot.gcf()

plt.scatter(ys, N1, marker=',')
#plt.scatter(ys, N2, marker=',')
plt.scatter(ys, N3, marker=',')
#plt.scatter(ys, N4, marker=',')
#plt.scatter(ys, N5, marker=',')
plt.scatter(ys, N6, marker=',')
#plt.scatter(ys, N7, marker=',')
fig.set_size_inches(w=40, h=15, forward=False)
#fig.savefig('test2png.png', dpi=200)
```

```python
lc_N1=[]
for idx, v in enumerate(N1):
    lc_N1.append([idx,v])

lc_N3=[]
for idx, v in enumerate(N3):
    lc_N3.append([idx,v])

lc_N6=[]
for idx, v in enumerate(N6):
    lc_N6.append([idx,v])
    
dataset = []
dataset.append(lc_N1)
dataset.append(lc_N3)
dataset.append(lc_N6)
dataset=pd.DataFrame(sum(dataset,[]))

dataset#,len(dataset),dataset[:10],dataset[-10:]
```

```python
from sklearn.cluster import KMeans, SpectralClustering, DBSCAN
import matplotlib.pyplot  as plt
import seaborn as sns

# create model and prediction
model = KMeans(n_clusters=3, algorithm='auto')

model.fit(dataset)
predict = pd.DataFrame(model.predict(dataset))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([dataset,predict],axis=1)

r
```

```python
sns.pairplot(r,hue='predict')
plt.show()
```
